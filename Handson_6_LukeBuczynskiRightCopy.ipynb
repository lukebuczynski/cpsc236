## 12.1 Champion Counter
Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

Italy	        4       1934, 1938, 1982, 2006

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,Jos√© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
### CODE HERE ###
FILENAME = "world_cup_champions.txt"
champions = {}
import csv

def readFile():
    win = 0
    with open(FILENAME, "r") as file:
        reader = csv.reader(file)
        wins = []
        
        for row in reader:
            if row[1] == "Country":
                continue
            elif row[0] == "Year":
                continue
            elif row[1] in champions:
                n= champions[row[1]][0]
                l= champions[row[1]][1]
                l.append(row[0])
                champions[row[1]] = [n+1, l]
            else:
                wins = [1, [row[0]]]
                champions[row[1]] = wins
    
def main():
    readFile()
    print("Country\t\tWins\tYears")
    print("=======\t\t====\t=====")
    for key in champions.keys():
        print(str(key) + "\t\t" + str(champions[key][0]) + "\t" + str(",".join(champions[key][1])))
    
if __name__ == "__main__":
    main()
````

## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
### CODE HERE ###
FILENAME = "monthly_sales.txt"
sales = {}
import csv 

def commandMenu():
    print("\nCOMMAND MENU")
    print("view - View sales for specified month")
    print("edit - Edit sales for specified month")
    print("totals - View sales summary for year")
    print("exit - Exit program") 

def readFile():
    with open(FILENAME, "r") as file:
        for line in file:
            sale = file.readlines()
        
        for i in range(len(sale)):
            month = sale[i]
            keys = month.split("	")
            key = keys[0]
            num = keys[1]
            
            sales[key] = num #num
    salesList = list(sales.keys())

    return salesList

def writeFile(salesList):
    with open(FILENAME, "w") as file:
        for i in range(len(salesList)):
            print(salesList[i] + "	" + sales[salesList[i]])
            file.write(str(salesList[i]) + "	" + str(sales[salesList[i]]) + "\n")
        
def commandView(salesList):
    mon = str(input("Three-Letter-Month: "))
    if mon in sales:
        print("Sales amount for " + mon + " is " + sales[mon])
    else:
        print("\nIncorrect Three-Letter-Month!")
    print()

def commandTotals(salesList):
    yearly = 0
    salesList = readFile()
    for i in range(0, len(salesList)):
        item = int(sales[salesList[i]])
        yearly += item
        
    print("\nYearly Total: " + str(yearly))
    avg = round((yearly/12), 2)
    print("Monthly Total: " + str(avg))

def commandEdit(salesList):
    index = 0
    while index == 0:
        abr = input("\nThree-letter Month: ")
        for i in range(0, len(salesList)+1):
            if salesList[i] == abr:
                index = i
                break
            else:
                index = 0
        if index == 0:
            print("Incorrect Three-Letter month!")
        else:
            sale = input("Sales: ")
    
            sales[abr] = sale
            print("Sales amount for " + abr + " has been modified!")
    
            writeFile(salesList)
            
def main():
    salesList = readFile()
    com = "cont"
    
    while com != "exit":
        commandMenu()
        com = str(input("\nWhich command do you want to do?: ").lower())
        if com == "view":
            commandView(salesList)
        elif com == "totals":
            commandTotals(salesList)
        elif com == "edit":
            commandEdit(salesList)
        elif com == "exit":
            break
        else:
            print("Wrong command!")
            
    print("Bye!")
    
if __name__ == "__main__":
    main()
#print("Total: {:, .2f").format(total)
````
