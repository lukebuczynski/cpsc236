## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell
Prime Number Checker

Please enter an integer between 1 and 5000: 5
5 is a prime number.

Try again? (y/n): y

Please enter an integer between 1 and 5000: 6
6 is NOT a prime number.
It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 3 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
### CODE HERE ###
print("Prime Number Checker")

def validInt():
    num = int(input("\nPlease enter an integer between 1 and 5000: "))
    while num <= 1 or num >= 5000:
        num = int(input("ERROR! Please enter an integer between 1 and 5000: "))
    return num
def outPut(num, fact, factor):
    if fact != 2:
        print(num, " is not a prime number!")
        print("It has " + str(fact) + " factors: ", factor)
    else:
        print(num, " is a prime number!")
def numFactors(num):
    factors = 1
    for i in range(1, num):
        if (num % i) == 0:
            factors += 1
    return factors
def findFactors(num):
    factors = []
    for i in range(1, num+1):
        if (num % i) == 0:
            factors.append(i)
    return factors
    
def main():
    cont = "y"
    allFactors = []
    while cont == "y":
        number = validInt()
        fact = numFactors(number)
        allFactors = findFactors(number)
        outPut(number, fact, allFactors)
        
        cont = str(input("\nTry again? (y/n): ").lower())
        while cont != "y" and cont != "n":
            cont = int(input("ERROR! Please input y/n: "))
            
    print("\nBye!")
    
if __name__=="__main__":
    main()
````

### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
### CODE HERE ###
def displayMenu():
    print("\nCOMMAND MENU")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add - Add a contact")
    print("del - Delete a contact")
    print("exit - Exit program")

def commandList(contactList):
    if len(contactList) == 0:
        print("There are no movies in the list!")
    else:
        i = 1
        for num in contactList:
            print(str(i) + ". " + str(num[0]))
            i += 1
        print()
        
def commandView(contactList):
    number = int(input("Number: "))
    if number < 1 or number > len(contactList):
        print("Invalid movie number.\n")
    else:
        number -= 1
        print("Name: " + contactList[number][0])
        print("Email: " + str(contactList[number][1]))
        print("Phone: " + str(contactList[number][2]))
    
def commandAdd(contactList):
    name = str(input("Name: "))
    email = str(input("Email: "))
    phone = str(input("Phone: "))
    contact = []
    contact.append(name)
    contact.append(email)
    contact.append(phone)
    contactList.append(contact)
    print(name, "was added")
    
def commandDel(contactList):
    number = int(input("Number: "))
    if number < 1 or number > len(contactList):
        print("Invalid movie number.\n")
    else:
        contactList = contactList.pop(number-1)
        print(contactList[0] + " was deleted.\n")
def commandExit():
    return 1
    

def main():
    print("CONTACT MANAGER")
    contactList = [["John Billy", "johnbilly@gmail.com", "724-333-444"],
                  ["Billy John", "billyjohn@gmail.com", "663-441-3213"]]
    displayMenu()
    while True:
        command = str(input("\nCommand: ").lower())
        if command == "list":
            commandList(contactList)
        elif command == "view":
            commandView(contactList)
        elif command == "add":
            commandAdd(contactList)
        elif command == "del":
            commandDel(contactList)
        elif command == "exit":
            break
        else:
            command = str(input("\nInvalid command, try again: "))
    print("Bye!")
    
if __name__ == "__main__":
    main()
````
