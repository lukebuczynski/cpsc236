## 7.1 Pig Dice Rules
Create a program that reads a list of rules from a file and displays them.

```powershell
*	Turn ends when player rolls a 1 or chooses to hold.

*	If you hold, you save all points earned during the turn.
```

### Specifications:
- Use the attached text file named `pig_dice_rules.txt` in the corresponding Directory.
- Your program should read the text file and display it on the console.

## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell
The Wizard Inventroy Program

COMMAND MENU
walk - Walk down the path show - Show all items drop - Drop an item
exit - Exit program

Command: walk
While walking down a path, you see a scroll of uncursing. Do you want to grab it? (y/n): y
You picked up a scroll of uncursing.

Command: walk
While walking down a path, you see an unknown potion. Do you want to grab it? (y/n): y
You can't carry any more items. Drop something first.

Command: show
1.	a wooden staff
2.	a scroll of invisibility
3.	a crossbow
4.	a scroll of uncursing

Command: drop Number: 3
You dropped a crossbow.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying. Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry four items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
### CODE HERE ###
import random

ALLITEMS = "wizard_all_items.txt"
INVENTORY = "wizard_inventory.txt"

def commandMenu():
    print("\nCOMMAND MENU")
    print("Walk - walk down the path!")
    print("Show - show all item!")
    print("Drop - drop an item!")
    print("Exit - exit program!")
    
def commandWalk(userInv):
    with open(ALLITEMS, "r") as file:
        listItems = file.readlines()
        
    randomNum = random.randint(0, len(listItems))
    randomItem = listItems[randomNum-1]
        
    choice = str(input("\nWhile walking down a path, you see " + str(randomItem) + ". Do you want to grab it? (y/n): ")).lower()
    while choice != "y" and choice != "n":
        choice = str(input("\nERROR! Please input (y/n): ")).lower()
    if choice == "y" and len(userInv) < 4:
        userInv.append(randomItem)

        with open(INVENTORY, "w") as f:
            for item in userInv:
                f.write(item)
        print("You picked up " + str(randomItem))
    elif choice == "y" and len(userInv) > 3:
        print("\nYou can't carry anymore items!")
    else:
        print("You leave the item behind...")
        
def commandShow():
    #use a loop to read each line of the file
    i = 1
    print("\nThe wizards inventory is: ")
    with open(INVENTORY, "r") as file:
        for line in file:
            print(str(i) + ". " + str(line), end="")
            i+=1
        print()
        
def commandDrop(userInv):
    commandShow()
    dropNum = int(input("\nWhich item do you want to drop? (1-4): "))
    while dropNum < 1 and dropNum > 4:
        dropNum = int(input("\nERROR! Please input a number (1-4): "))
    print("You dropped " + str(userInv[dropNum-1]) + "!")
    userInv.pop(dropNum-1)
    with open(INVENTORY, "w") as f:
        for item in userInv:
            f.write(item)
    
def main():
    com = " "
    userInv = []
    with open(INVENTORY, "r") as f:
        for line in f:
            userInv.append(line)
    print("The wizard inventory program!")
    while com != "exit":
        commandMenu()
        com = str(input("\nCommand: ").lower())
        while com != "walk" and com != "drop" and com != "show" and com != "exit":
            com = str(input("ERROR! Please enter a valid command: ").lower())
        if com == "walk":
            commandWalk(userInv)
        elif com == "show":
            commandShow()
        elif com == "drop":
            commandDrop(userInv)
        else:
            break
    print("\nBye!")
    
if __name__ == "__main__":
    main()
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:
```powershell

COMMAND MENU
```

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesn’t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
### CODE HERE ###
import csv
FILENAME = "monthly_sales.csv"

def commandMenu():
    print("\nmonthly - View monthly sales yearly - View yearly sumary edit   - Edit sales for a month exit   - Exit program")

def readCSV():
    monthly = []
    with open(FILENAME, newline="") as file:
        reader = csv.reader(file)
        for row in reader:
            monthly.append(row)
    return monthly

def writeCSV(salesList):
    with open(FILENAME, "w", newline = "") as file:
        writer = csv.writer(file)
        writer.writerows(salesList)
        
def commandMonthly(salesList):
    print()
    for i in range(len(salesList)):
        sales = salesList[i]
        print(sales[0] + " - " + sales[1])
    print()

def commandYearly(salesList):
    yearly = 0
    salesList = readCSV()
    for i in range(0, len(salesList)):
        item = int(salesList[i][1])
        yearly += item
        i+=1
        
    print("\nYearly Total: " + str(yearly))
    avg = round((yearly/12), 2)
    print("Monthly Total: " + str(avg))

def commandEdit(salesList):
    index = 0
    while index == 0:
        abr = input("\nThree-letter Month: ")
        for i in range(0, len(salesList)):
            if salesList[i][0] == abr:
                index = i
                break
            else:
                index = 0
        if index == 0:
            print("Incorrect Three-Letter month!")
        else:
            sales = input("Sales: ")
    
            salesList[index][1] = sales
            print("Sales amount for " + abr + " has been modified!")
    
            writeCSV(salesList)
            
def main():
    salesList = readCSV()
    com = "cont"
    
    while com != "exit":
        commandMenu()
        com = str(input("Which command do you want to do?: ").lower())
        if com == "monthly":
            commandMonthly(salesList)
        elif com == "yearly":
            commandYearly(salesList)
        elif com == "edit":
            commandEdit(salesList)
        elif com == "exit":
            break
        else:
            print("Wrong command!")
            
    print("Bye!")
    
if __name__ == "__main__":
    main()
````
