## 8.1 Tip Calculator
Add exception handling to a Tip Calculator program.

### Console:
```powershell
Tip Calculator

INPUT
Cost of meal: ten
Must be valid decimal number. Please try again. 
Cost of meal: -10
Must be greater than 0. Please try again. 
Cost of meal: 52.31
Tip percent:	17.5
Must be valid integer. Please try again. 
Tip percent:	20

OUTPUT
Cost of meal:   52.31 
Tip percent:    20%
Tip amount:     10.46
Total amount:   62.77
```

### Specifications:
- The program should accept **decimal** entries like 52.31 and 15.5 for the cost of the meal.
- The program should accept **integer** entries like 15, 20, 25 for the tip percent.
- The program should validate both user entries. That way, the user can’t crash the program by entering invalid data.
- The program should only accept numbers that are **greater than 0**.
- The program should round results to a maximum of two decimal places.


````python
### CODE HERE ###
def getMealCost():
    try:
        mealCost = float(input("Cost Of Meal: "))
        while mealCost < 0:
            mealCost = float(input("Must be a value greater than 0, please try again: "))
        return mealCost
    except ValueError:
        print("Invalid Decimal Type! Please try again")
        return 0
    
def getTip():
    try:
        tipPer = int(input("Tip Percent: "))
        while tipPer != 15 and tipPer != 20 and tipPer != 25:
            tipPer = int(input("\nERROR! Please enter a correct tip percent: "))
        return tipPer
    except ValueError:
        print("Invalid Integer Type! Please try again")
        return 0

def main():
    print("Tip Calculator!")
    print("\nINPUT")
    meal = getMealCost()
    while meal == 0:
        meal = getMealCost()
    tip = getTip()
    while tip == 0:
        tip = getTip()
        
    tipPer=tip/100
    tipAmt=round((meal*tipPer), 2)
    totalAmt=round((meal+tipAmt), 2)
    
    print("\nOutput")
    print("Cost of Meal: "+str(meal))
    print("Tip: "+str(tip)+"%")
    print("Tip amount: $"+str(tipAmt))
    print("Total amount: $"+str(totalAmt)+"\n")
        
    print("Bye!")
if __name__ == "__main__":
    main()
````

### Specifications:
- This program should read the text file named `wizard_all_items.txt` that contains all the items a wizard can carry.
- When the user selects the walk command, the program should randomly pick one of the items that were read from the text file and give the user the option to grab it.
- The current items that the wizard is carrying should be saved in an inventory file. Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry four items at a time. For the drop command, display an error message if the user enters an invalid integer or an integer that doesn’t correspond with an item.
- Handle all exceptions that might occur so that the user can’t cause the program to crash. If the all items file is missing, display an appropriate error message and exit the program.
- If the inventory file is missing, display an appropriate error message and continue with an empty inventory for the user. That way, the program will write a new inventory file when the user adds items to the inventory.

````python
### CODE HERE ###W
import random


ALLITEMS = "wizard_all_items.txt"
INVENTORY = "wizard_inventory.txt"
userInv = []

def commandMenu():
    print("\nCOMMAND MENU")
    print("Walk - walk down the path!")
    print("Show - show all item!")
    print("Drop - drop an item!")
    print("Exit - exit program!")
    
def commandWalk(userInv):
    try:
        with open(ALLITEMS, "r") as file:
            listItems = file.readlines()
        
        randomNum = random.randint(0, len(listItems))
        randomItem = listItems[randomNum-1]
        
        choice = str(input("\nWhile walking down a path, you see " + str(randomItem) + ". Do you want to grab it? (y/n): ")).lower()
        while choice != "y" and choice != "n":
            choice = str(input("\nERROR! Please input (y/n): ")).lower()
        if choice == "y" and len(userInv) < 4:
            userInv.append(randomItem)

            with open(INVENTORY, "w") as f:
                for item in userInv:
                    f.write(item)
            print("You picked up " + str(randomItem))
        elif choice == "y" and len(userInv) > 3:
            print("\nYou can't carry anymore items!")
        else:
            print("You leave the item behind...")
        return "cont"
    except FileNotFoundError:
        print("\nERROR! The file cannot be found, make sure you have the wizard_all_items.txt file!")
        return "exit"
             
        
def commandShow():
    #use a loop to read each line of the file
    try:
        i = 1
        print("\nThe wizards inventory is: ")
        with open(INVENTORY, "r") as file:
            for line in file:
                print(str(i) + ". " + str(line), end="")
                i+=1
            print()
    except FileNotFoundError:
        print("\nERROR! The file " + INVENTORY + " is not found!")
        
def commandDrop(userInv):
    try:
        commandShow()
        dropNum = int(input("\nWhich item do you want to drop? (1-4): "))
        while dropNum < 1 and dropNum > 4:
            dropNum = int(input("\nERROR! Please input a number (1-4): "))
        print("You dropped " + str(userInv[dropNum-1]) + "!")
        userInv.pop(dropNum-1)
        with open(INVENTORY, "w") as f:
            for item in userInv:
                f.write(item)
    except IndexError:
        print("Not enough items in the wizard's inventory!")
    except ValueError:
        print("Invalid Item Number")

def main():
    com = " "
    try:
        with open(INVENTORY, "r") as f:
            for line in f:
                userInv.append(line)
        while com != "exit":
            commandMenu()
            com = str(input("\nCommand: ").lower())
            while com != "walk" and com != "drop" and com != "show" and com != "exit":
                com = str(input("ERROR! Please enter a valid command: ").lower())
            if com == "walk":
                com = commandWalk(userInv)
            elif com == "show":
                commandShow()
            elif com == "drop":
                commandDrop(userInv)
            else:
                break
    except FileNotFoundError:
        print("Could not find anything in file " + INVENTORY + ", wizard is starting with an empty inventory")
        with open(INVENTORY, "w") as f:
            pass

    print("\nBye!")
    
if __name__ == "__main__":
    main()
````
